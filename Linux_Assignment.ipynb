# Linux


1. What is Linux ?
   - Linux is a free and open-source operating system (OS) kernel, widely used as the foundation for various operating systems and software.

2. What is the difference between Hard Link & Soft Link ?
  - Hard Link : A hard link is like creating another name for the same file, it points directly to the file’s data on the disk. This means that both the original file and the hard link share the same content, and if one is changed, the other reflects those changes. Even if the original file is deleted, the data remains safe and can still be accessed through the hard link. However, hard links cannot be used for folders or across different storage drives or partitions.
  - Command to create hard link : ln original_file hard_link_name

  - Soft Link : A soft link is more like a shortcut. It doesn’t point to the file’s actual data, but instead to the file name or path. If the original file is moved or deleted, the soft link becomes broken and won’t work anymore. The advantage of soft links is that they can link to folders and also work across different file systems.
  -  Command to create soft link : ln -s original_file soft_link_name

3. What is a Kernel in Linux ?
   -   The kernel in Linux is the core part of the operating system. It acts like a bridge between the computer hardware and the software (applications). The kernel manages everything in the system, such as memory, processes (programs running), devices (keyboard, mouse, hard disk), and files.The kernel is open-source, which means developers can view, modify, and improve its code. This is one of the main reasons why Linux is so flexible, powerful, and used in many devices—from mobile phones to servers and supercomputers.

4. How do you create a user account ?  
   - To create a user account in Linux, we can use the "adduser" command with root or sudo privileges. Open the terminal and type : sudo adduser username.
   - Ex : sudo adduser vaishnavi , this will create a new user named vaishnavi with a home directory. /home/vaishnavi.

5. What is the ‘grep’ command used for in Linux ?
   - The 'grep' command in Linux is used to search for specific text or patterns within files or output. It stands for "Global Regular Expression Print". we can use 'grep' to quickly find lines that contain a particular word, phrase, or pattern.
   - For example, to search for the word “error” in a file called 'log.txt', you would use : grep "error" log.txt command .
   - This will display all lines from 'log.txt' that contain the word "error",  'grep' is a powerful tool often used in combination with other commands to filter and analyze text data.
